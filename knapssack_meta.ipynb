```python
import numpy as np
import math
```

```python
n=5
m=15
w = [12, 2, 1, 1, 4]
v = [4, 2, 1, 2, 10]

dvar = [0 ,1 ,1, 0, 1]
```

```python
class DNA:
    def __init__(self, dv=[]):
        if len(dv)==0:
            self.dvar = np.random.randint(0,2,n)
        else:
            self.dvar = dv
        self.fitness = None
        self.total_weight = None

    def calc_fitness(self):
        self.fitness = np.sum(self.dvar * np.array(v))
        self.total_weight = np.sum(self.dvar * np.array(w))
        if self.total_weight>m:
            violate = self.total_weight-m
            self.fitness = self.fitness - 4*violate
    
    def mutate(self):
        for i in range(n):
            if np.random.rand()<0.2:
                self.dvar[i] = 1-self.dvar[i]

```

```python
def crossover(dna1, dna2):
    mid = np.random.randint(0,n)
    new_dna = DNA(dv=np.concatenate([dna1.dvar[0:mid], dna2.dvar[mid:n]]))
    return new_dna
```

```python
abc = DNA()
abc.calc_fitness()
print(abc.dvar)
print(abc.fitness)
print(abc.total_weight)
```

```python
def fitness_curve(fit):
    return (1/2)* pow(fit,2)
```

```python
fitness_curve(2)
```

```python
# init population
pop_size = 20
population = []
for i in range(pop_size):
    population.append(DNA())
    population[i].calc_fitness()
```

```python
import math
pool = []
for i in range(pop_size):
    for j in range(math.floor(fitness_curve(population[i].fitness))):
        pool.append(i)
```

```python
num_of_it = 100
# best_dna = DNA()
best_fitness = -10000

for it in range(num_of_it):
    print("Iteration", it)
    new_population = []
    pool = []
    for i in range(pop_size):
        print(population[i].fitness)
        for j in range(math.floor(fitness_curve(population[i].fitness))):
            pool.append(i)
    for i in range(pop_size):
        id1 = np.random.choice(pool)
        id2 = np.random.choice(pool)
        new_dna = crossover(population[id1], population[id2])
        new_dna.mutate()
        new_dna.calc_fitness()
        if new_dna.fitness > best_fitness:
            best_fitness = new_dna.fitness
            best_dna = new_dna
            print(best_fitness)
        new_population.append(new_dna)
    population = new_population
    
best_dna.calc_fitness()
print(best_dna.fitness)
print(best_dna.dvar)
```
